```python
licenses=pd.read_csv(path+"licenses.csv")
licenses=licenses.iloc[:,1:]
biz_owners=pd.read_csv(path+"business_owners.csv")
biz_owners=biz_owners.iloc[:,1:]
```

```python
licenses.head()
```

```python
biz_owners.head()
```

```python
# Merge the licenses and biz_owners table on account
licenses_owners = licenses.merge(biz_owners, on='account')

# Group the results by title then count the number of accounts
counted_df = licenses_owners.groupby('title').agg({'account':'count'})

# Sort the counted_df in desending order
sorted_df = counted_df.sort_values(by='account', ascending=False)

# Use .head() method to print the first few rows of sorted_df
sorted_df.head()
```

```python
cal=pd.read_csv(path+"cta_calendar.csv")
cal=cal.iloc[:,1:]
ridership=pd.read_csv(path+"cta_ridership.csv")
ridership=ridership.iloc[:,1:]
stations=pd.read_csv(path+"stations.csv")
stations=stations.iloc[:,1:]
```

```python
# Merge the ridership, cal, and stations tables
ridership_cal_stations = ridership.merge(cal, on=['year','month','day']) \
            				.merge(stations,on="station_id")
ridership_cal_stations.head()
```

```python
# Merge the ridership, cal, and stations tables
ridership_cal_stations = ridership.merge(cal, on=['year','month','day']) \
							.merge(stations, on='station_id')

# Create a filter to filter ridership_cal_stations
filter_criteria = ((ridership_cal_stations['month'] == 7) 
                   & (ridership_cal_stations['day_type'] == 'Weekday') 
                   & (ridership_cal_stations['station_name'] == 'Wilson'))

# Use .loc and the filter to select for rides
print(ridership_cal_stations.loc[filter_criteria, 'rides'].sum())
```

```python
land_use=pd.read_csv(path+"land_use.csv")
land_use=land_use.iloc[:,1:]
land_use.head()
```

```python
movies=pd.read_csv(path+"movies.csv")
financials=pd.read_csv(path+"financials.csv")

movies=movies.iloc[:,1:]
financials=financials.iloc[:,1:]
```

# Left Join

```python
# Merge movies and financials with a left join
movies_financials = movies.merge(financials,on="id",how="left")
movies_financials.head()
```

```python
# Count the number of rows in the budget column that are missing
number_of_missing_fin = movies_financials['budget'].isnull().sum()

# Print the number of movies missing financials
print(number_of_missing_fin)
```

# Self Join

```python
crews = pd.read_csv(path+"crews.csv")
crews = crews.iloc[:,1:]
crews.head()
```

```python
# Merge the crews table to itself
crews_self_merged = crews.merge(crews,on="id",how="inner",suffixes=('_dir','_crew'))
crews_self_merged.head()
```

```python
# Merge the crews table to itself
crews_self_merged = crews.merge(crews, on='id', how='inner',
                                suffixes=('_dir','_crew'))

# Create a Boolean index to select the appropriate rows
boolean_filter = ((crews_self_merged['job_dir'] == 'Director') & 
                  (crews_self_merged['job_crew'] != 'Director'))
direct_crews = crews_self_merged[boolean_filter]
direct_crews.head()
```

```python
financials = pd.read_csv(path+"financials.csv")
financials = financials.iloc[:,1:]
sequels = pd.read_csv(path+"sequels.csv")
sequels = sequels.iloc[:,1:]
```

```python
# Merge sequels and financials on index id
sequels_fin = sequels.merge(financials, on='id', how='left')

# Merge sequels and financials on index id
sequels_fin = sequels.merge(financials, on='id', how='left')

# Self merge with suffixes as inner join with left on sequel and right on id
orig_seq = sequels_fin.merge(sequels_fin, how='inner', left_on='sequel', 
                             right_on='id', right_index=True,
                             suffixes=('_org','_seq'))

# Add calculation to subtract revenue_org from revenue_seq 
orig_seq['diff'] = orig_seq['revenue_seq'] - orig_seq['revenue_org']

# Merge sequels and financials on index id
sequels_fin = sequels.merge(financials, on='id', how='left')

# Self merge with suffixes as inner join with left on sequel and right on id
orig_seq = sequels_fin.merge(sequels_fin, how='inner', left_on='sequel', 
                             right_on='id', right_index=True,
                             suffixes=('_org','_seq'))

# Add calculation to subtract revenue_org from revenue_seq 
orig_seq['diff'] = orig_seq['revenue_seq'] - orig_seq['revenue_org']

# Select the title_org, title_seq, and diff 
titles_diff = orig_seq[['title_org','title_seq','diff']]
```

```python
# Print the first rows of the sorted titles_diff
titles_diff.sort_values('diff', ascending=False).head()
```

# merge_ordered()

```python
gdp = pd.read_csv(path+"WorldBank_GDP.csv")
gdp = gdp.iloc[:,1:]

sp500 = pd.read_csv(path+"SP500.csv")
```

```python
# Use merge_ordered() to merge gdp and sp500 on year and date
gdp_sp500 = pd.merge_ordered(gdp, sp500, left_on='Year', right_on='Date', how='left')
# Print gdp_sp500
gdp_sp500.head()
```

```python
# Use merge_ordered() to merge gdp and sp500 on year and date
gdp_sp500 = pd.merge_ordered(gdp, sp500, left_on='Year', right_on='Date', how='left',fill_method="ffill")
# Print gdp_sp500
gdp_sp500.head()
```

```python
# Subset the gdp and returns columns
gdp_returns = gdp_sp500[['GDP','Returns']]

# Print gdp_returns correlation
print(gdp_returns.corr())
```
